```python
from numpy import genfromtxt
import numpy as np
```

```python
asset_data = {'AA':[],'AAL':[],'AAPL':[],'TSLA':[],'W5000':[],'GSPC':[]}

for asset_key in asset_data:
    asset_data[asset_key] = genfromtxt('./data/'+asset_key+'.csv', skip_header=1, usecols=(4), delimiter=',',dtype=None, encoding=None)

print(asset_data)
```

```python
asset_matrix = np.array([asset_data['AA'],asset_data['AAL'],asset_data['AAPL'],asset_data['TSLA'],asset_data['W5000'],asset_data['GSPC']])
market_summary = np.mean(asset_matrix, axis=0)

print(market_summary)
```

```python
market_cov = {'AA':None,'AAL':None,'AAPL':None,'TSLA':None,'W5000':None,'GSPC':None}

for asset_key in market_cov:
    market_cov[asset_key] = np.cov(asset_data[asset_key],market_summary)
    market_cov[asset_key] = market_cov[asset_key][0][1]

test_qubo = np.diag(list(market_cov.values()))
test_qubo = -1*test_qubo/6
test_qubo[test_qubo == 0] = 100
print(test_qubo)
#print(market_cov)
```

```python
asset_beta = {'AA':None,'AAL':None,'AAPL':None,'TSLA':None,'W5000':None,'GSPC':None}

for asset_key in asset_beta:
    asset_beta[asset_key] = market_cov[asset_key][0][1]/np.var(market_summary)

print(asset_beta)
```

```python
asset_cov = {}

for asset_key_a in asset_data:
    for asset_key_b in asset_data:
        asset_cov[asset_key_a+'_'+asset_key_b] = np.cov(asset_data[asset_key_a],asset_data[asset_key_b])
        
print(asset_cov)
```

```python
print(asset_cov['AAPL_AA'])
```

```python
print(asset_cov['AA_AAPL'])
```

```python
asset_return = {'AA':None,'AAL':None,'AAPL':None,'TSLA':None,'W5000':None,'GSPC':None}
for asset_key in asset_return:
    asset_return[asset_key] = asset_data[asset_key][-1] - asset_data[asset_key][0]

all_asset_return = 0
for asset_key in asset_return:
    all_asset_return += asset_return[asset_key]
    
print(asset_return)
print(all_asset_return)
```

```python
risk_free_return = 2.63
print(risk_free_return)
```

```python
asset_std = {'AA':None,'AAL':None,'AAPL':None,'TSLA':None,'W5000':None,'GSPC':None}

for asset_key in asset_std:
    asset_std[asset_key] = np.std(asset_data[asset_key])

print(asset_std)
```

```python
all_asset_std = 0
for asset_key in asset_std:
    all_asset_std +=  asset_std[asset_key]
    
all_asset_std = np.sqrt(all_asset_std)

print(all_asset_std)
```

```python
sharpe_ratio_all_assets = (all_asset_return - risk_free_return)/all_asset_std

print(sharpe_ratio_all_assets)
```
